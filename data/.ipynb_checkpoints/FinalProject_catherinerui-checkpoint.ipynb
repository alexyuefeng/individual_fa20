```python
%matplotlib inline

import pandas as pd
import geopandas as gpd
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (17, 5)
plt.rcParams.update({'font.size': 16})
from mpl_toolkits.axes_grid1 import make_axes_locatable
import seaborn as sns

import shapely.geometry as shp

import sklearn.neighbors as skn
import sklearn.metrics as skm

import warnings
warnings.filterwarnings('ignore')

pd.options.display.max_rows = 10

#improve resolution
#comment this line if erroring on your machine/screen
%config InlineBackend.figure_format ='retina'
import patsy
import statsmodels.api as sm
```

```python
### Question: What is the relationship between yelp ratings, number of yelp reviews, provision of phone number and zip code of a park
```

```python
df1 = gpd.read_file('parks_datasd.geojson')
df2 = pd.read_csv('yelp_SD_parks.csv')
```

```python
df2['phonenumber'] = df2['phone'].apply(lambda x: 0 if np.isnan(x) else 1)
df2['zipcode'] = df2['address'].apply(lambda x: x.split(',')[-1].strip())
df2['zipcode'].unique()
```

```python
df2[df2['zipcode']=='Baja California Mexico']['address']
```

```python
df2[df2['zipcode']=='CA']['address']
```

```python
df2 = df2.drop([312,556,226,314,747]).reset_index(drop = True)
```

```python
df2['zipcode'] = df2.zipcode.apply(lambda x: np.int(x[-5:]))
```

```python
geo2 = gpd.GeoDataFrame(
    df2, geometry=gpd.points_from_xy(df2.longitude, df2.latitude))
```

```python
geo2.columns
```

```python
geo2 = geo2[['name', 'address', 'review_count','zipcode','phonenumber','geometry','rating']]
geo2.head()
```

```python
len(geo2)
```

```python
geo2.describe() 
```

```python
 #Q1 – 1.5×IQR or above Q3 + 1.5×IQR
q1 = np.percentile(geo2['review_count'],25)
q2 = np.percentile(geo2['review_count'],75)
lower = q1 - 1.5*(q2-q1)
upper = q1 = 1.5*(q2-q1)
geo2 = geo2[(geo2['review_count']>=lower)&(geo2['review_count']<= upper)]
```

```python
outcome, predictors = patsy.dmatrices('rating ~ review_count', geo2)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
print(res.summary())
```

```python
outcome, predictors = patsy.dmatrices('rating ~ zipcode', geo2)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
print(res.summary())
```

```python
#plot original data
sns.scatterplot(x='zipcode', y='rating', data=geo2)

#generate model fit line
xs = np.arange(geo2['zipcode'].min(), geo2['zipcode'].max())
ys = 0.00003982 * xs +  0.3748
plt.plot(xs, ys, '--k', linewidth=4, label='Model')

plt.xlabel('Yelp Rating')
plt.ylabel('Inspection Score')
plt.title('Relationship between Yelp Rating and Inspection Score')
plt.legend();
```

```python
outcome, predictors = patsy.dmatrices('rating ~ phonenumber', geo2)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
print(res.summary())
```

```python
outcome, predictors = patsy.dmatrices('review_count ~ zipcode', geo2)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
print(res.summary())
```

```python
outcome, predictors = patsy.dmatrices('review_count ~ phonenumber', geo2)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
print(res.summary())
```

```python
#plot original data
sns.scatterplot(x='phonenumber', y='review_count', data=geo2)

#generate model fit line
xs = np.arange(0,1,0.001)
ys = 15.1913 * xs + 18.8898
plt.plot(xs, ys, '--k', linewidth=4, label='Model')

plt.xlabel('provision of phone number')
plt.ylabel('review count')
plt.title('Relationship between review count and provision of phone number')
plt.legend();
```
