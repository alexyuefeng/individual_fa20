# Data Cleaning

The primary goals of data cleaning for this project will be to:

- **(1)** Discard data that is not needed for analysis
- **(2)** Rename columns as needed for readability
- **(3)** Merge Data to make analysis easier

### Step 0 : Go through Steps #1-3 for Data Cleaning listed

```python
yelp_SD_parks = yelp_SD_parks[["name", "address"]]

# zip-code is always the last 5 characters given any address.
yelp_SD_parks["zip_code"] = yelp_SD_parks["address"].str[-5:]
yelp_SD_parks
```

```python
# To be safe, convert zip_codes from str to int, and anything that can't
# gets marked to be dropped, because parks with zip-codes are important 
# for later analysis.
def zip_code_check(input_str):
    try:
        output = int(input_str)
        return output
    except ValueError:
        output = "drop"
        return output
    except TypeError:
        output = "drop"
        return output
    
yelp_SD_parks["zip_code"] = yelp_SD_parks["zip_code"].apply(zip_code_check)
```

```python
# Then we drop all rows that contain the value "drop" from column "zip_code"
drop_rows_indexes = yelp_SD_parks[yelp_SD_parks["zip_code"] == "drop"].index

yelp_SD_parks.drop(drop_rows_indexes, inplace=True)
yelp_SD_parks
```

```python
yelp_SD_reviews = yelp_SD_reviews[["id", "rating"]]
yelp_SD_reviews
```

```python
# Getting the average rating values from yelp reviews
average_ratings = yelp_SD_reviews.groupby(["id"])["rating"].mean()
average_yelp_SD_reviews = pd.Series.to_frame(average_ratings)
average_yelp_SD_reviews = average_yelp_SD_reviews.rename(columns={"rating":"avg_rating"})

# Merging the average_yelp_SD_reviews with yelp_SD_reviews into my review df
review_df = pd.merge(average_yelp_SD_reviews, yelp_SD_reviews, on="id")

# Dropping the "rating" column from df since it's un-necessary at this point
review_df = review_df.drop(columns="rating")

# Renaming the "id" column from df to "name" for readability
review_df = review_df.rename(columns={"id":"name"})

# Dropping duplicate names from "names" because name:avg_rating should be 1:1
review_df = review_df.drop_duplicates(subset="name")

# Creating a primary df that'll have review_df merge with yelp_SD_parks
df = pd.merge(review_df, yelp_SD_parks, on="name")

# Dropping rows w/ duplicate names even if they have different locations
# This is because parks w/ duplicate names but different addresses could
# very much be due to the user-input error or a park being large enough to
# have multiple addresses.
df = df.drop_duplicates(subset="name")

# Reset the index to have nice order
df.reset_index()
df
```

```python
# Creating a new dataframe that will comprise of the average rating of yelp 
# reviewed parks by zip code. Then it'll be sorted in ascending order.
zip_rating = df.groupby(["zip_code"])["avg_rating"].mean()
avg_zip_rating = pd.Series.to_frame(zip_rating)
avg_zip_rating = avg_zip_rating.rename(columns={"avg_rating":"avg_zip_rating"})
avg_zip_rating.reset_index(inplace=True)

avg_zip_rating = avg_zip_rating.sort_values(by="avg_zip_rating",ascending=True)
avg_zip_rating.reset_index()
```

```python
# Medians amongst the avg_rating & avg_zip_rating
avg_rating_median = df["avg_rating"].median()
avg_zip_rating_median = avg_zip_rating["avg_zip_rating"].median()

print(avg_rating_median, avg_zip_rating_median)
```

Additionally, I will be dropping additional parks based on their review_count.
If a park has below 10 reviews, I will be not considering it because I can only conclude off of the existing data that there is definitely not a sufficient sample of reviewers for the rating of the park to have any weight.

```python
# Review_count is int64 type.
#print(type(yelp_SD_parks["review_count"].iloc[0]))

def review_counter(int_input):
    if int_input < 10:
        output = 0
    else:
        output = int_input
    return output

# Apply review_counter to yelp_SD_parks
yelp_SD_parks["review_count"] = yelp_SD_parks["review_count"].apply(review_counter)

drop_rows_indexes = yelp_SD_parks[yelp_SD_parks["review_count"] == 0].index

yelp_SD_parks.drop(drop_rows_indexes, inplace=True)
yelp_SD_parks
```
